```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix, plot_confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
def generate_accuracy_and_heatmap(model, x, y):
    ac = metrics.accuracy_score(y, model.predict(x))
    f_score = metrics.f1_score(y, model.predict(x))
    
    print("Accuracy: ", ac)
    print("F1 score: ", f_score)
    print("\n")
    print(pd.crosstab(pd.Series(model.predict(x), name='Predicted'),
                     pd.Series(y['Outcome'], name='Actual')))
    return 1
```

```python
# separating X and y variables
X = df.drop(['target'], axis=1)
y = df['target']
```

```python
X.shape, y.shape
```

```python
# separating the dataset into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

```python
# classifier
svm_clf = SVC(kernel="linear")
svm_clf_baseline_model = svm_clf.fit(X_train, y_train)
```

```python
# checking accuracy using the created function
generate_accuracy_and_heatmap(svm_clf_baseline_model,X_test, y_test)
```

```python
# accuracy before balancing the dataset
# print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
# print(classification_report(y_test, y_pred))
```

```python
# Feature selection
from sklearn.feature_selection import RFE
from sklearn.feature_selection import RFECV
```

```python
rfe = RFE(estimator=svm_clf, step=1)
rfe = rfe.fit(X_train, y_train)
```

```python
selected_rfe_features = pd.DataFrame({'Feature':list(X_train.columns),
                                    'Ranking':rfe.ranking_})
selected_rfe_features.sort_values(by='Ranking')
```

```python
# using the transform function to transform existing training data 
# to transformed 6 columns
x_train_rfe = rfe.transform(X_train)
x_test_rfe = rfe.transform(X_test)
```

```python
# transformed data
x_train_rfe[0:3]
```

```python
# fitting the model with the 6 transformed features
svm_rfe_model = svm_clf.fit(x_train_rfe, y_train)
```

```python
generate_accuracy_and_heatmap(svm_rfe_model, x_test_rfe, y_test)
```

```python
# accuracy before balancing the dataset
# print("Accuracy:",metrics.accuracy_score(y_test, y_pred_rfe))
```

```python
#print(classification_report(y_test, y_pred_rfe))
```

```python
# recursive feature elimination with cross validation
rfecv = RFECV(estimator=svm_clf, step=1, cv=5, scoring='accuracy')
rfecv = rfecv.fit(X_train, y_train)
print('Optimal number of features: ', rfecv.n_features_)
print('Best features: ', X_train.columns[rfecv.support_])
```

```python
# grid scores for each feature
rfecv.grid_scores_
```

```python
# plotting the graph based on cross validation of selected features
plt.figure()
plt.xlabel("Number of selected features")
plt.ylabel("Cross validation score of number of selected features")
plt.plot(range(1, len(rfecv.grid_scores_) + 1), rfecv.grid_scores_)
```

```python
# transforming newly selected features
x_train_rfecv = rfecv.transform(X_train)
x_test_rfecv = rfecv.transform(X_test)
```

```python
# fitting the model with new features
svm_rfecv_model = svm_clf.fit(x_train_rfecv, y_train)
```

```python
# checking accuracy
generate_accuracy_and_heatmap(svm_rfecv_model, x_test_rfecv, y_test)
```
